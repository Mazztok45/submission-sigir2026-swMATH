```julia
using HTTP
using JSON3
using Base.Filesystem

"""
    APIClient

Base class for API interactions with common functionality.
"""
mutable struct APIClient
    base_url::String
    headers::Dict{String, String}

    function APIClient(base_url::String, headers::Union{Dict, Nothing}=nothing)
        if headers === nothing
            headers = Dict("Content-Type" => "application/json")
        end
        new(base_url, headers)
    end
end

"""
    get(client::APIClient, endpoint::String, params::Union{Dict, Nothing}=nothing)

Make a GET request to the API.
"""
function get(client::APIClient, endpoint::String, params::Union{Dict, Nothing}=nothing)::Dict
    url = "$(rstrip(client.base_url, '/'))/$(lstrip(endpoint, '/'))"

    response = HTTP.get(url; headers=client.headers, query=params)

    if response.status != 200
        error("HTTP Error: $(response.status)")
    end

    return JSON3.read(String(response.body))
end

"""
    download_file(client::APIClient, url::String, output_path::String, chunk_size::Int=8192)

Download a file from a URL.
"""
function download_file(client::APIClient, url::String, output_path::String, chunk_size::Int=8192)::String
    response = HTTP.get(url; headers=client.headers)

    if response.status != 200
        error("HTTP Error: $(response.status)")
    end

    open(output_path, "w") do f
        write(f, response.body)
    end

    return output_path
end

"""
    InfomaniakDriveClient

Client for Infomaniak Drive API.
"""
mutable struct InfomaniakDriveClient
    api_client::APIClient
    drive_id::String

    function InfomaniakDriveClient(token::String, drive_id::String)
        headers = Dict("Authorization" => "Bearer $token")
        api_client = APIClient("https://api.infomaniak.com/3/drive", headers)
        new(api_client, drive_id)
    end
end

"""
    list_folder_contents(client::InfomaniakDriveClient, folder_id::String)

List all files in a folder.
"""
function list_folder_contents(client::InfomaniakDriveClient, folder_id::String)::Vector
    endpoint = "$(client.drive_id)/files/$folder_id/files"
    response = get(client.api_client, endpoint)
    return get(response, "data", [])
end

"""
    download_folder_files(client::InfomaniakDriveClient, folder_id::String, output_dir::String=".")

Download all files from a folder.
"""
function download_folder_files(client::InfomaniakDriveClient, folder_id::String, output_dir::String=".")::Vector{String}
    files = list_folder_contents(client, folder_id)
    downloaded_files = String[]

    for item in files
        if item["type"] == "file"
            file_id = item["id"]
            file_name = item["name"]
            download_url = "https://api.infomaniak.com/2/drive/$(client.drive_id)/files/$file_id/download"

            println("Downloading $file_name...")
            output_path = joinpath(output_dir, file_name)
            download_file(client.api_client, download_url, output_path)
            push!(downloaded_files, output_path)
            println("✓ Downloaded $file_name")
        end
    end

    return downloaded_files
end

"""
    download_file_by_id(client::InfomaniakDriveClient, file_id::String, output_name::String)

Download a specific file by ID.
"""
function download_file_by_id(client::InfomaniakDriveClient, file_id::String, output_name::String)::String
    download_url = "https://api.infomaniak.com/2/drive/$(client.drive_id)/files/$file_id/download"
    println("Downloading $output_name...")
    return download_file(client.api_client, download_url, output_name)
end

# Initialize the client
infomaniak = InfomaniakDriveClient(
    "Fi9Aaadhibox6l50TcideMovSUo8JtaKc30i_yP-XJFcf_db9HL_BX9KQ0HZN41H8u53Pu6tGRq3PGx5",
    "705884"
)
```

```julia
"""
Memory-Optimized swMATH2GNN Data Processing Pipeline
Designed to handle large datasets with limited RAM
"""

__precompile__(false)


# Navigate to project directory
if basename(pwd()) != "swMATH2GNN"
    cd("..")
end

using DataFrames
using SparseArrays
using MLLabelUtils
using StatsBase
using Arrow
using Serialization
using CSV
using HTTP
using JSON3
using URIs
using Logging
using Tables

# Include data extraction modules
include("../swMATH2GNN/src/extracting_data/extract_software_metadata.jl")
include("../swMATH2GNN/src/extracting_data/extract_article_metadata.jl")

import .DataReaderswMATH: generate_software_dataframe
import .DataReaderszbMATH: extract_articles_metadata

# ============================================================================
# CONFIGURATION
# ============================================================================
const BATCH_SIZE = 500  # Smaller batches for memory safety
const DATA_DIR = "GNN_Julia"
const FINAL_TABLE = joinpath(DATA_DIR, "df_arrow")
const CROSSREF_DIR = "crossref_titles"
const REQUEST_DELAY = 0.08
const COMBINE_CHUNK_SIZE = 5  # Combine only 5 batches at a time

mkpath(DATA_DIR)
mkpath(CROSSREF_DIR)

# ============================================================================
# MEMORY UTILITIES
# ============================================================================

function force_gc()
    """Force garbage collection and print memory stats"""
    GC.gc(true)
    GC.gc(true)  # Call twice for thorough cleanup
end

function log_memory_usage(stage::String)
    """Log current memory usage"""
    stats = Base.gc_num()
    println("[$stage] Memory: $(round(stats.total_allocd / 1024^3, digits=2)) GB allocated")
end

# ============================================================================
# PHASE 1: DATA EXTRACTION & BATCHING (MEMORY OPTIMIZED)
# ============================================================================

function extract_and_batch_articles()
    if isfile(FINAL_TABLE)
        println("✓ Final table exists, skipping extraction")
        return true
    end

    println("\n" * "="^70)
    println("PHASE 1: EXTRACTING & BATCHING ARTICLES (MEMORY OPTIMIZED)")
    println("="^70)

    # Get total count without loading all data
    println("Extracting articles metadata...")
    articles_list_dict = extract_articles_metadata()
    total = length(articles_list_dict)
    println("✓ Found $total articles")

    batch_num = 0
    successful_batches = 0

    for start_idx in 1:BATCH_SIZE:total
        end_idx = min(start_idx + BATCH_SIZE - 1, total)
        batch_num += 1

        try
            # Extract only current batch
            batch = articles_list_dict[start_idx:end_idx]

            # Process batch
            node_features = [dic[:msc] for dic in batch]
            software = [dic[:software] for dic in batch]
            paper_ids = [dic[:id] for dic in batch]
            titles = [dic[:title] for dic in batch]
            dois = [haskey(dic, :doi) ? dic[:doi] : "no doi" for dic in batch]
            references = [dic[:ref_ids] for dic in batch]

            df_batch = DataFrame(
                node_features = node_features,
                software = software,
                paper_id = paper_ids,
                title = titles,
                doi = dois,
                references = references
            )

            batch_file = joinpath(DATA_DIR, "batch_$(lpad(batch_num, 5, '0')).arrow")
            Arrow.write(batch_file, df_batch)
            successful_batches += 1

            println("✓ Batch $batch_num saved ($(nrow(df_batch)) rows) - Progress: $(end_idx)/$(total)")

            # Aggressive memory cleanup
            node_features = nothing
            software = nothing
            paper_ids = nothing
            titles = nothing
            dois = nothing
            references = nothing
            df_batch = nothing
            batch = nothing

            # Force GC every batch
            if batch_num % 2 == 0
                force_gc()
            end

        catch e
            println("✗ Error batch $batch_num: $e")
            @warn "Stack trace:" exception=(e, catch_backtrace())
            continue
        end
    end

    # Clear source data
    articles_list_dict = nothing
    force_gc()
    log_memory_usage("After extraction")

    println("\n✓ Successfully created $successful_batches batches")
    return successful_batches > 0
end

# ============================================================================
# PHASE 1B: COMBINE BATCHES INCREMENTALLY
# ============================================================================

function combine_batches_incrementally()
    """Combine batch files incrementally to avoid memory overflow"""

    println("\n📦 Combining batch files incrementally...")
    batch_files = sort(filter(f -> endswith(f, ".arrow") && startswith(f, "batch_"),
                              readdir(DATA_DIR)))

    if isempty(batch_files)
        println("✗ No batch files found!")
        return false
    end

    total_batches = length(batch_files)
    println("Found $total_batches batch files to combine")

    # Combine in chunks to avoid memory overflow
    temp_combined_files = []

    for chunk_start in 1:COMBINE_CHUNK_SIZE:total_batches
        chunk_end = min(chunk_start + COMBINE_CHUNK_SIZE - 1, total_batches)
        chunk_files = batch_files[chunk_start:chunk_end]

        println("  Combining chunk $(div(chunk_start-1, COMBINE_CHUNK_SIZE)+1): batches $chunk_start-$chunk_end")

        # Load first file in chunk
        df_chunk = DataFrame(Arrow.Table(joinpath(DATA_DIR, chunk_files[1])))

        # Append remaining files in chunk
        for batch_file in chunk_files[2:end]
            batch_df = DataFrame(Arrow.Table(joinpath(DATA_DIR, batch_file)))
            df_chunk = vcat(df_chunk, batch_df)
            batch_df = nothing
            force_gc()
        end

        # Save temporary combined chunk
        temp_file = joinpath(DATA_DIR, "temp_combined_$(lpad(chunk_start, 5, '0')).arrow")
        Arrow.write(temp_file, df_chunk)
        push!(temp_combined_files, temp_file)

        println("    ✓ Chunk saved: $(nrow(df_chunk)) rows")
        df_chunk = nothing
        force_gc()
    end

    # Now combine all temporary files
    println("\n  Final combination of $(length(temp_combined_files)) chunks...")
    df_final = DataFrame(Arrow.Table(temp_combined_files[1]))

    for temp_file in temp_combined_files[2:end]
        temp_df = DataFrame(Arrow.Table(temp_file))
        df_final = vcat(df_final, temp_df)
        temp_df = nothing
        force_gc()
    end

    # Save final table
    Arrow.write(FINAL_TABLE, df_final)
    println("✓ Final table saved: $(nrow(df_final)) rows")

    total_rows = nrow(df_final)
    df_final = nothing
    force_gc()

    # Cleanup batch and temp files
    println("\n🧹 Cleaning up temporary files...")
    for batch_file in batch_files
        rm(joinpath(DATA_DIR, batch_file), force=true)
    end
    for temp_file in temp_combined_files
        rm(temp_file, force=true)
    end

    log_memory_usage("After combining batches")
    return true
end

# ============================================================================
# PHASE 2: CROSSREF TITLE ENRICHMENT (STREAMING)
# ============================================================================

function fetch_title_from_crossref(doi, request_delay)
    try
        sleep(request_delay)
        response = HTTP.get(
            "https://api.crossref.org/works/$(URIs.URI(doi))",
            ["User-Agent" => "swMATH2GNN/1.0"];
            readtimeout=30,
            retry_non_idempotent=true
        )

        if response.status == 200
            data = JSON3.read(response.body)
            return get(data["message"]["title"], 1, missing)
        end
    catch e
        @warn "Error fetching $doi: $e"
    end
    return missing
end

function process_single_doi(doi, crossref_dir, request_delay)
    filename = joinpath(crossref_dir, replace("$(doi).txt", "/" => "_"))

    if !isfile(filename)
        title = fetch_title_from_crossref(doi, request_delay)
        if title !== missing
            try
                open(filename, "w") do f
                    write(f, title)
                end
            catch e
                @warn "Error writing file for $doi: $e"
            end
        end
    end
    return nothing
end

function enrich_titles_from_crossref_streaming()
    """Stream through dataframe to enrich titles without loading all into memory"""
    println("⚠️ SKIPPING Crossref enrichment (user requested)")
    arrow_table = Arrow.Table(FINAL_TABLE)
    Arrow.write(
        joinpath(DATA_DIR, "titles_enriched.arrow"),
        Tables.table(collect(arrow_table.title), header=["title"])
    )
    return true

    if isfile(joinpath(DATA_DIR, "titles_enriched.arrow"))
        println("✓ Enriched titles exist, skipping Crossref fetch")
        return true
    end

    println("\n" * "="^70)
    println("PHASE 2: ENRICHING TITLES FROM CROSSREF (STREAMING)")
    println("="^70)

    # Stream through Arrow file in chunks
    arrow_table = Arrow.Table(FINAL_TABLE)
    total_rows = length(arrow_table.title)

    # Identify DOIs needing enrichment (stream through)
    dois_to_fetch = String[]
    indices_to_update = Int[]

    println("Scanning for titles needing enrichment...")
    for i in 1:total_rows
        if arrow_table.title[i] == "Not available" &&
           !isnothing(arrow_table.doi[i]) &&
           arrow_table.doi[i] != "no doi"
            push!(dois_to_fetch, arrow_table.doi[i])
            push!(indices_to_update, i)
        end

        if i % 10000 == 0
            println("  Scanned $i/$total_rows rows")
        end
    end

    if isempty(dois_to_fetch)
        println("✓ No titles need enrichment")
        # Just copy titles as-is
        Arrow.write(
            joinpath(DATA_DIR, "titles_enriched.arrow"),
            Tables.table(collect(arrow_table.title), header=["title"])
        )
        return true
    end

    println("Found $(length(dois_to_fetch)) titles to enrich")

    # Fetch titles from Crossref
    for (idx, doi) in enumerate(dois_to_fetch)
        process_single_doi(doi, CROSSREF_DIR, REQUEST_DELAY)

        if idx % 50 == 0
            println("  Progress: $idx/$(length(dois_to_fetch)) DOIs processed")
            force_gc()
        end
    end

    # Build enriched titles array
    println("\nBuilding enriched titles array...")
    enriched_titles = collect(arrow_table.title)  # Start with original
    l_files = readdir(CROSSREF_DIR)

    for (idx, doi) in zip(indices_to_update, dois_to_fetch)
        str_file = replace(doi, "https://doi.org/" => "") |> x -> replace(x, "/" => "_")

        if "$(str_file).txt" in l_files
            try
                title_content = read("$(CROSSREF_DIR)/$(str_file).txt", String)
                enriched_titles[idx] = strip(title_content)
            catch e
                @warn "Error reading title for $doi: $e"
            end
        end

        if idx % 1000 == 0 && idx in indices_to_update
            println("  Updated $(findfirst(==(idx), indices_to_update))/$(length(indices_to_update)) titles")
        end
    end

    # Save enriched titles
    Arrow.write(
        joinpath(DATA_DIR, "titles_enriched.arrow"),
        Tables.table(enriched_titles, header=["title"])
    )

    println("✓ Enriched titles saved")
    enriched_titles = nothing
    force_gc()
    log_memory_usage("After title enrichment")
    return true
end

# ============================================================================
# PHASE 3: MSC ENCODING (MEMORY EFFICIENT)
# ============================================================================

function encode_msc_sparse()
    """Create MSC encoding keeping it sparse to save memory"""

    println("\n" * "="^70)
    println("PHASE 3: ENCODING MSC CODES (SPARSE)")
    println("="^70)

    if isfile("sparse_one_hot.jls")
        println("✓ MSC encoding exists")
        return true
    end

    # Load data in streaming fashion
    arrow_table = Arrow.Table(FINAL_TABLE)
    enriched_titles = Arrow.Table(joinpath(DATA_DIR, "titles_enriched.arrow")).title

    # Filter: keep only rows with valid titles
    valid_indices = findall(x -> x != "Not available", enriched_titles)
    println("Valid papers: $(length(valid_indices)) / $(length(enriched_titles))")

    # Extract 2-digit MSC codes (streaming)
    println("Extracting MSC codes...")
    all_msc_2digit = Set{String}()
    msc_by_paper = Vector{Vector{String}}()

    for idx in valid_indices
        node_features = arrow_table.node_features[idx]
        msc_2 = unique([string(x[1:2]) for x in node_features])
        push!(msc_by_paper, msc_2)
        union!(all_msc_2digit, msc_2)

        if length(msc_by_paper) % 10000 == 0
            println("  Processed $(length(msc_by_paper)) papers")
        end
    end

    # Create encoding
    maj_cat = sort(collect(all_msc_2digit))
    value_to_index = Dict(value => i for (i, value) in enumerate(maj_cat))

    println("Creating sparse one-hot matrix: $(length(msc_by_paper)) x $(length(maj_cat))")

    # Build sparse matrix incrementally
    I_indices = Int[]
    J_indices = Int[]

    for (i, msc_list) in enumerate(msc_by_paper)
        for msc_code in msc_list
            push!(I_indices, i)
            push!(J_indices, value_to_index[msc_code])
        end

        if i % 10000 == 0
            println("  Built sparse entries for $i papers")
        end
    end

    sparse_one_hot = sparse(I_indices, J_indices, ones(Bool, length(I_indices)),
                           length(msc_by_paper), length(maj_cat))

    # Save as sparse (much more memory efficient)
    serialize("sparse_one_hot.jls", sparse_one_hot)
    serialize("valid_indices.jls", valid_indices)

    println("✓ Sparse MSC encoding saved: $(size(sparse_one_hot))")
    println("  Sparsity: $(round(100 * (1 - nnz(sparse_one_hot) / prod(size(sparse_one_hot))), digits=2))% zeros")

    force_gc()
    log_memory_usage("After MSC encoding")
    return true
end

# ============================================================================
# PHASE 4: GROUPED DATA (STREAMING)
# ============================================================================


function prepare_grouped_data_streaming()
    """Group data by paper_id using streaming to minimize memory"""

    println("\n" * "="^70)
    println("PHASE 4: PREPARING GROUPED DATA (STREAMING)")
    println("="^70)

    if isfile("grouped_data_by_paper_id.csv")
        println("✓ Grouped data exists")
        return true
    end

    # Load necessary data
    arrow_table = Arrow.Table(FINAL_TABLE)
    enriched_titles = Arrow.Table(joinpath(DATA_DIR, "titles_enriched.arrow")).title
    valid_indices = deserialize("valid_indices.jls")

    println("Processing $(length(valid_indices)) valid papers...")

    # Build grouped data dictionary (paper_id -> data)
    grouped_dict = Dict{Int, Dict{Symbol, Any}}()

    for idx in valid_indices
        paper_id = arrow_table.paper_id[idx]
        software_raw = arrow_table.software[idx]
        node_features = arrow_table.node_features[idx]
        title = enriched_titles[idx]

        # CRITICAL FIX: Convert software to vector of integers
        # Check what type software_raw is
        if software_raw isa String
            # If it's a string, parse it
            software = parse.(Int, split(software_raw, ","))
        elseif software_raw isa AbstractVector
            # If it's already a vector, convert to integers
            software = collect(software_raw)
        else
            # Otherwise just use it as-is
            software = software_raw
        end

        if !haskey(grouped_dict, paper_id)
            grouped_dict[paper_id] = Dict(
                :software => Set{Int}(),
                :msc => Set{String}(),
                :title => title
            )
        end

        # Add software
        union!(grouped_dict[paper_id][:software], software)

        # Add 2-digit MSC codes
        msc_2 = [string(x[1:2]) for x in node_features]
        union!(grouped_dict[paper_id][:msc], msc_2)

        if length(grouped_dict) % 5000 == 0
            println("  Grouped $(length(grouped_dict)) unique papers")
        end
    end

    println("✓ Created groups for $(length(grouped_dict)) unique papers")

    # Convert to DataFrame
    grouped_df = DataFrame(
        paper_id = Int[],
        merged_msc_codes_2 = String[],
        merged_software = String[],
        title = String[]
    )

    for (paper_id, data) in grouped_dict
        push!(grouped_df, (
            paper_id = paper_id,
            merged_msc_codes_2 = join(sort(collect(data[:msc])), ","),
            merged_software = join(sort(collect(data[:software])), ","),
            title = data[:title]
        ))
    end

    # Sort by paper_id
    sort!(grouped_df, :paper_id)

    CSV.write("grouped_data_by_paper_id.csv", grouped_df)
    println("✓ Grouped data saved: $(nrow(grouped_df)) rows")

    grouped_dict = nothing
    grouped_df = nothing
    force_gc()
    log_memory_usage("After grouping")
    return true
end

#
# ============================================================================
# PHASE 5: SOFTWARE EDGES & MULTI-HOT (FIXED)
# ============================================================================

function create_software_edges_and_multihot()
    """Create software graph edges and multi-hot encoding"""

    println("\n" * "="^70)
    println("PHASE 5: SOFTWARE EDGES & MULTI-HOT ENCODING")
    println("="^70)

    # Load grouped data
    grouped_df = DataFrame(CSV.File("grouped_data_by_paper_id.csv"))
    software_arrays = map(x -> parse.(Int, split(x, ",")), grouped_df.merged_software)

    # Create label_to_index mapping FIRST (needed for both edges and mapping)
    all_software = sort(unique(vcat(software_arrays...)))
    label_to_index = Dict(label => i for (i, label) in enumerate(all_software))

    # Save label_to_index for later use
    serialize("label_to_index.jls", label_to_index)

    # Software edges
    if !isfile("software_graph2.edgelist")
        println("Creating software edges...")
        DataReaderswMATH.df = DataFrame()
        software_df = generate_software_dataframe()

        related_edges = sortslices(
            hcat(software_df.id, map(x -> x[:id], software_df.related_software)),
            dims=1,
            by=x->(x[1],x[2]),
            rev=false
        )

        unique_software = unique(reduce(vcat, software_arrays))

        software_edges = reduce(
            hcat,
            filter(
                x -> x[1] in unique_software && x[2] in unique_software,
                eachrow(related_edges)
            )
        )'

        CSV.write("software_graph2.edgelist", Tables.table(software_edges); header=false)
        println("✓ Software edges saved")

        # Save software_df for Phase 6 to avoid calling generate_software_dataframe() again
        serialize("software_df_temp.jls", software_df)
    else
        println("✓ Software edges exist")
    end

    # Multi-hot encoding
    if !isfile("sparse_multi_hot.jls")
        println("\nCreating sparse multi-hot encoding...")

        # Build sparse matrix
        I_indices = Int[]
        J_indices = Int[]

        for (i, software_list) in enumerate(software_arrays)
            for software in software_list
                push!(I_indices, i)
                push!(J_indices, label_to_index[software])
            end

            if i % 5000 == 0
                println("  Processed $i/$(length(software_arrays)) papers")
            end
        end

        sparse_multi_hot = sparse(I_indices, J_indices, ones(Bool, length(I_indices)),
                                 length(software_arrays), length(all_software))

        serialize("sparse_multi_hot.jls", sparse_multi_hot)
        println("✓ Sparse multi-hot saved: $(size(sparse_multi_hot))")
        println("  Sparsity: $(round(100 * (1 - nnz(sparse_multi_hot) / prod(size(sparse_multi_hot))), digits=2))% zeros")
    else
        println("✓ Multi-hot encoding exists")
    end

    force_gc()
    log_memory_usage("After software processing")
    return true
end

# ============================================================================
# PHASE 6: SOFTWARE MAPPING (FIXED)
# ============================================================================

function create_software_mapping()
    """Create software-to-articles mapping CSV"""

    println("\n" * "="^70)
    println("PHASE 6: CREATING SOFTWARE MAPPING")
    println("="^70)

    if isfile("software_articles_mapping.csv")
        println("✓ Software mapping exists")
        return true
    end

    # Load the saved software_df from Phase 5 instead of regenerating
    println("Loading software metadata...")
    if isfile("software_df_temp.jls")
        software_df = deserialize("software_df_temp.jls")
    else
        software_df = generate_software_dataframe()
    end

    # Load the saved label_to_index mapping
    label_to_index = deserialize("label_to_index.jls")

    # Create index column using the mapping
    software_df[!, :index] = [Base.get(label_to_index, id, missing) for id in software_df[!, :id]]

    # Report statistics
    has_index_count = count(!ismissing, software_df[!, :index])
    println("Software IDs with indices: $has_index_count")
    println("Software IDs without indices: $(count(ismissing, software_df[!, :index]))")
    println("Total software rows: $(nrow(software_df))")

    # Save mapping
    CSV.write("software_articles_mapping.csv",
              software_df[!, [:id, :index, :name, :keywords]])

    # Cleanup temp file
    if isfile("software_df_temp.jls")
        rm("software_df_temp.jls", force=true)
    end

    println("✓ Software mapping saved")
    return true
end
# ============================================================================
# MAIN PIPELINE
# ============================================================================

function main()
    println("\n" * "█"^70)
    println("swMATH2GNN - MEMORY-OPTIMIZED PIPELINE")
    println("█"^70)

    try
        # Phase 1: Extract & batch
        println("\n🔄 Starting Phase 1: Data Extraction...")
        if !extract_and_batch_articles()
            println("✗ Phase 1 failed")
            return
        end

        # Phase 1B: Combine batches
        println("\n🔄 Starting Phase 1B: Combining Batches...")
        if !combine_batches_incrementally()
            println("✗ Phase 1B failed")
            return
        end

        # Phase 2: Enrich titles
        println("\n🔄 Starting Phase 2: Title Enrichment...")
        if !enrich_titles_from_crossref_streaming()
            println("✗ Phase 2 failed")
            return
        end

        # Phase 3: Encode MSC
        println("\n🔄 Starting Phase 3: MSC Encoding...")
        if !encode_msc_sparse()
            println("✗ Phase 3 failed")
            return
        end

        # Phase 4: Group data
        println("\n🔄 Starting Phase 4: Grouping Data...")
        if !prepare_grouped_data_streaming()
            println("✗ Phase 4 failed")
            return
        end

        # Phase 5: Software edges & multi-hot
        println("\n🔄 Starting Phase 5: Software Processing...")
        if !create_software_edges_and_multihot()
            println("✗ Phase 5 failed")
            return
        end

        # Phase 6: Software mapping
        println("\n🔄 Starting Phase 6: Software Mapping...")
        if !create_software_mapping()
            println("✗ Phase 6 failed")
            return
        end

        println("\n" * "█"^70)
        println("✓ ALL PHASES COMPLETE")
        println("█"^70)
        println("\nOutputs created:")
        println("  - $(FINAL_TABLE)")
        println("  - sparse_one_hot.jls (MSC encoding)")
        println("  - sparse_multi_hot.jls (Software encoding)")
        println("  - grouped_data_by_paper_id.csv")
        println("  - software_graph2.edgelist")
        println("  - software_articles_mapping.csv")
        println("\n💡 All matrices kept sparse to minimize memory usage")

    catch e
        println("\n✗ Pipeline failed with error:")
        println(e)
        @warn "Stack trace:" exception=(e, catch_backtrace())
    finally
        force_gc()
        log_memory_usage("Final")
    end
end

# ============================================================================
# EXECUTION
# ============================================================================

if abspath(PROGRAM_FILE) == @__FILE__
    main()
end
```
